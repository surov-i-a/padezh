```python
import numpy as np
import pandas as pd

from numpy import dot, array, zeros, linspace, mod, sqrt, mean, std, exp, sin, cos, pi, imag, real, angle, mod
from numpy.linalg import norm

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
font = {'size'   : 12}
mpl.rc('font', **font)
```

```python
from navec import Navec
path = 'navec_hudlit_v1_12B_500K_300d_100q.tar'
#path = '/home/surov/Library/navec_news_v1_1B_250K_300d_100q.tar'
model = Navec.load(path)
```

```python
dt = pd.read_csv('data_6x2891.csv', index_col=0)
dt
```

```python
# Форматирую данные для функций sklearn. N = число использованных строк датафрейма
# LabelEncoding. Целевая величина y=0 для родительного, 1 для предложного, 2 для винительного и т.д.
N = 1000;
N = len(dt)
x, y = [], []
for n in range(N):
    for k in range(6):
        x.append(model[dt.iloc[n][k]])
        y.append(k)

x = np.array(x)
y = np.array(y)

x.shape, y.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=2024, test_size=0.25)
y_train
```

# Классификация методом КNN

```python
# Число соседей по умолчанию 5
knn = KNeighborsClassifier(n_neighbors=5, weights='uniform')
knn.fit(x_train, y_train)

y_preds_knn = knn.predict(x_test)
print(classification_report(y_test, y_preds_knn))
```

```python
# То же самое с весами-расстояниями
knn = KNeighborsClassifier(n_neighbors=5, weights='distance')
knn.fit(x_train, y_train)

y_preds_knn = knn.predict(x_test)
print(classification_report(y_test, y_preds_knn))
```

## <font color='red'>Вывод</font>
Оба решения существенно хуже бейзлайна (0.73)

# Подбор оптимального n_neighbors

```python
f1list = []
for k in range(10,500,20):
    knn = KNeighborsClassifier(n_neighbors=k, weights='uniform', n_jobs=-1)
    knn.fit(x_train, y_train)
    y_preds_knn = knn.predict(x_test)
    f1 = classification_report(y_test, y_preds_knn, output_dict=True)['weighted avg']['f1-score']
    f1list.append([k,f1])
f1list = array(f1list)
```

```python
fig,ax = plt.subplots(1,1,figsize=(8,4))
ax.plot(f1list[:,0], f1list[:,1], marker='o')
ax.grid(True, ls='--');
ax.set_xlabel('Число соседей')
ax.set_ylabel('Средняя F1');
ax.set_title('KNN, weights=uniform');
```

```python
# То же самое с весами-расстояниями
f1list = []
for k in range(10,500,20):
    knn = KNeighborsClassifier(n_neighbors=k, weights='distance')
    knn.fit(x_train, y_train)
    y_preds_knn = knn.predict(x_test)
    f1 = classification_report(y_test, y_preds_knn, output_dict=True)['weighted avg']['f1-score']
    f1list.append([k,f1])
f1list = array(f1list)
```

```python
fig,ax = plt.subplots(1,1,figsize=(8,4))
ax.plot(f1list[:,0], f1list[:,1], marker='o')
ax.grid(True, ls='--');
ax.set_xlabel('Число соседей')
ax.set_ylabel('Средняя F1');
ax.set_title('KNN, weights=distance');
```

## <font color='red'>Вывод</font>
Наилучшая точность при weights=uniform с числом соседей примерно 50-200. При этом средняя F1 = 71% всё же ниже бейзлайна (0.73)

## Метрики для оптимального решения

```python
knn = KNeighborsClassifier(n_neighbors=170, weights='uniform', algorithm='brute')
knn.fit(x_train, y_train)

y_preds_knn = knn.predict(x_test)
print(classification_report(y_test, y_preds_knn))
```

```python
conmat = confusion_matrix(y_test, y_preds_knn)
conmat
```

```python
fig,ax = plt.subplots(1,1,figsize=(5,5))
ax.imshow(conmat, cmap='Reds', vmin=0);
ax.set_ylabel('Истина')
ax.set_xlabel('Модель');
```

## <font color='red'>Вывод</font>
Крайние классы определяются лучше чем средние (ощущение что значения Y обладают метрикой возникло отсюда).
Основная ошибка набирается за счёт смешивания именительного и винительного падежей, в которых формы слов почти всегда совпадают. Остальные падежи можно достаточно достоверно определять по данным модели word2vec.

### Снижение размерности методом NeighborhoodComponentsAnalysis

```python
from sklearn.neighbors import NeighborhoodComponentsAnalysis
```

```python
# Нахождение наилучшей плоскости для различения классов
nca = NeighborhoodComponentsAnalysis(n_components=3);
nca.fit(x_train, y_train)
```

```python
# Сохраняю т.к. считается минут 5. 
# Получил на старой версии scikit (~0.24). После обновления до 1.2.1 стала вылетать
import pickle
'''with open('nca_dim3.pkl', 'wb') as file:
    pickle.dump(nca, file)'''
    
with open('nca_dim3.pkl', 'rb') as file:
    nca = pickle.load(file)
```

```python
# Применение KNN к трансформированным данным

knn = KNeighborsClassifier(n_neighbors=70)
knn.fit(nca.transform(x_train), y_train)

y_preds_knn = knn.predict(nca.transform(x_test))
```

```python
print(classification_report(y_test, y_preds_knn))
```

```python
xdim3 = nca.transform(x)
```

```python
# Окраска классов в порядке столбцов датафрейма: 
# Родительный - голубой, предложный - синий, винительный - фиолетовый, именительный красный, дательный желтый, творительный зелёный.
cmap = mpl.colors.ListedColormap(['c','b','m','r','y','g'])
norm = mpl.colors.BoundaryNorm([0,1,2,3,4,5,6], 6)

fig, ax = plt.subplots(1,3,figsize=(18,6))
ax[0].scatter(xdim3[:, 0], xdim3[:, 1], c=y, s=1, cmap=cmap, norm=norm);
ax[1].scatter(xdim3[:, 0], xdim3[:, 2], c=y, s=1, cmap=cmap, norm=norm);
ax[2].scatter(xdim3[:, 1], xdim3[:, 2], c=y, s=1, cmap=cmap, norm=norm);

for a in ax: 
    a.grid(True, ls='--')
```

```python
conmat = confusion_matrix(y_test, y_preds_knn)
fig,ax = plt.subplots(1,1,figsize=(5,5))
ax.imshow(conmat, cmap='Greens', vmin=0);
ax.set_ylabel('Истина')
ax.set_xlabel('Модель');
```

## <font color='red'>Вывод</font>
Целевые классы различаются в трёхмерном (и двумерном) пространствах почти так же хорошо как и в исходном 300d: F1 мера 0.71 против 0.73. Матрица ошибок примерно такая же. 

Группировка цветов на первом графике показывает, что падежи русского языка в модели word2vec характеризуются определёнными геометрическими отношениями.

## <font color='red'>Вывод</font>
По сравнению с предыдущим точность практически без изменений, но второй метод выявил другую геометрическую структуру. Творительный падеж (зелёный) уникален по первой оси d1, остальные лежат в лоскости d2-d3. В этой плоскости предложный (синий), именительный (красный) и дательный (жёлтый) лежат в вершинах треугольника, а родительный (голубой) близко к его центру. Это расположение согласуется с конкретной теорией смысла.  Объяснять довольно долго, но это замечательно

```python
# Дополнительная отрисовка

dtp = pd.DataFrame(data=xdim3, columns=['d1','d2','d3'])
dtp['y'] = y

# Без винительного (фиотеловый)
dtp_no2 = dtp[dtp['y'] != 2]
# Без винительного и творительного (фиотеловый и зелёный)
dtp_no25 = dtp[(dtp['y'] != 2) & (dtp['y'] != 5)]
```
